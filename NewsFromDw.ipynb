```python
import feedparser
"""
Get all news from dw.com
"""
url="https://rss.dw.com/xml/rss-de-all"

news=feedparser.parse(url)

i = 0
for news in news.entries:
    i += 1
    print(i)
    print(news.title)
    print(news.link)
    print(news.summary)
    print("\n")
```
